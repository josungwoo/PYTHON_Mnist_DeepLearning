```python
import numpy as np
import matplotlib.pyplot as plt
import mnist.mnist
```

# 1. Configuration Parameters
 - 언더스코어는 derived parameter이므로 변경가능 하지 않음 !
 - derived parameter는 이미 정의된 다른 파라미터로부터 값이 생성되는 파라미터님

```python
DISPLAY_CNT = 12                                    # 디스플레이 대상 이미지 샘플 개수
FIG_ROW_CNT = 2                                     # 화면에 디스플레이의 행의 개수
_FIG_COL_CNT = int(DISPLAY_CNT / FIG_ROW_CNT)       # 한 행당 이미지 개수 (소수점이하 삭제)
_TOTAL_DISPLAY_CNT =  FIG_ROW_CNT * _FIG_COL_CNT    # 실제 디스플에할 이미지 개수
```

# 2. MNIST 데이터셋 로딩

```python
# MNIST  데이터셋 로딩
(x_train, d_train), (test_x, d_test) = mnist.mnist.load_mnist(
            flatten=True, normalize=False, one_hot_label=False)

# 훈련 데이터의 샘플 및 라벨의 개수
print(x_train.shape, d_train.shape)
# 테스트 데이터의 샘플 및 라벨의 개수
print(test_x.shape, d_test.shape)
```

# 3. 첫번째 train 샘플  살펴보기

```python
img = x_train[0]
label = d_train[0]
print(img.shape)
print(label)

img = img.reshape(28,28)
plt.imshow(img)
```

# 4. 훈련 샘플 디스플레이

```python
j = 1
for i in range(_TOTAL_DISPLAY_CNT):
    plt.subplot(FIG_ROW_CNT, _FIG_COL_CNT, j )
    img = x_train[i]
    plt.title(" Sample %d" % i)
    img = img.reshape(28,28)
    plt.imshow(img)
    j = j + 1
    
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9,
                    hspace=0.1, wspace=0.5)
plt.show()
```
